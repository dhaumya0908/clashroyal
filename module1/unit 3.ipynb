```python
#importing the requried moduel
# pip insatall matplotlib
import matplotlib.pyplot as plt
x = [3,2,5]
y = [1,2,3]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('My First Graph!!')
plt.show()
```

```python
import numpy as np
a = np.arange(0,10,0.1)
b = np.sin(a)

#figures in mmatplotlib using figure()
plt.figure(figsize=(10,8))
plt.plot(a,b)
plt.show()
```
